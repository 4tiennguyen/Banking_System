```python
connection = pymysql.connect(
    host='localhost',
    user='root',  # enter your username
    password='',  # enter your password 
    database='banking_system'
)

data_frames = {
    'Person': person_df,
    'Account': account_df,
    'Transaction': transactions_df,
    'Branch': branch_df,
    'Employee': employees_df
}
try:
    with connection.cursor() as cursor:
        for table_name, df in data_frames.items():
            for idx, row in df.iterrows():
                # Replace NaT values with None or a default date
                row = row.where(pd.notnull(row), None)  # Replaces NaT with None

                placeholders = ', '.join(['%s'] * len(row))
                columns = ', '.join(row.index)
                sql = f"INSERT INTO {table_name} ({columns}) VALUES ({placeholders})"
                cursor.execute(sql, tuple(row))

    connection.commit()
    print("Data inserted successfully.")

except Exception as e:
    print("Error:", e)
    connection.rollback()

finally:
    connection.close()

```

```python
import pymysql
import hashlib
import string
import secrets


with open('C:/Users/tttie/OneDrive/Documents/CS 257/simulatedData/person_data.json', 'r', encoding='utf-8') as json_file:
    person_data = json.load(json_file)
    person_df = pd.DataFrame(person_data)

# Establish a connection to the database
connection = pymysql.connect(
    host='localhost',
    user='root',
    password='',
    database='banking_system'
)

def generate_password():
    alphabet = string.ascii_letters + string.digits
    password = ''.join(secrets.choice(alphabet) for i in range(10))
    return password

def hash_password(password):
    hashed_password = hashlib.sha256(password.encode()).hexdigest()
    return hashed_password

# Assuming 'person_df' has the necessary data extracted from the JSON

# Iterate through each row in the DataFrame to insert data into the UserCredentials table
with connection.cursor() as cursor:
    for index, row in person_df.iterrows():
        person_id = row['PersonID']
        email = row['Email']
        password = generate_password()
        hashed_password = hash_password(password)

        print(f"Generated password for PersonID: {person_id} is: {password}")

        sql = f"INSERT INTO UserCredentials (PersonID, Email, Password) VALUES ('{person_id}', '{email}', '{hashed_password}')"

        try:
            cursor.execute(sql)
            connection.commit()
            print(f"Inserted UserCredential data for PersonID: {person_id} with a generated password.")
        except Exception as e:
            connection.rollback()
            print(f"Error occurred while inserting data for PersonID: {person_id}")
            print("Error:", e)
```
