## Employee Table

```python


# Create a list of roles and the corresponding number of employees for each role
roles = [
    {"Role": "Manager", "Count": 1},
    {"Role": "Teller", "Count": 3},
    {"Role": "Accountant", "Count": 2},
    {"Role": "Customer Service", "Count": 2},
    {"Role": "Assistant Manager", "Count": 1}
]

# Create an empty list to store employee records
employees = []

# Initialize PersonIDs
person_ids = list(range(1, 501))

# Shuffle the PersonIDs to randomize assignments
random.shuffle(person_ids)

# Assign employees to 5 branches with the specified roles
branch_count = 5  # Number of branches
for branch_id in range(1, branch_count + 1):
    for role in roles:
        for _ in range(role["Count"]):
            if person_ids:
                person_id = person_ids.pop()
                employees.append({
                    "EmployeeID": len(employees) + 1,
                    "Position": role["Role"],
                    "PersonID": person_id,
                    "BranchID": branch_id
                })

# Create a DataFrame of employees
employees_df = pd.DataFrame(employees)

# Display the employee DataFrame
employees_df
```

```python

# List to store transactions
transactions = []

# Map to store related transactions (withdrawal to deposit)
related_transactions = {}

# Function to generate random dates within a given range
def random_date(start_date, end_date):
    return start_date + relativedelta(
        hours=random.randint(8, 16),  # Random hour between 8 AM and 4 PM
        minutes=random.randint(0, 59),  # Random minute
    )

# Loop through each account in the 'accounts_df' DataFrame
for index, account in accounts_df.iterrows():
    # Convert 'DateOpen' to a datetime object, including the time component
    date_open = datetime.strptime(account['DateOpen'], '%Y-%m-%d %H:%M:%S')

    # Calculate the minimum TransactionDate based on the DateOpen of the account
    minimum_transaction_date = date_open

    # Determine the number of transactions for this account
    num_transactions = random.randint(1, 10)

    for _ in range(num_transactions):
        # Generate a random TransactionDate within the allowed range
        transaction_date = random_date(minimum_transaction_date, datetime(2023, 1, 1))

        # Determine the transaction type (Deposit, Withdrawal, or Transfer)
        transaction_type = random.choice(['Deposit', 'Withdrawal', 'Transfer'])

        # Determine the transaction amount
        if transaction_type == 'Withdrawal':
            max_withdrawal_amount = account['CurrentBalance']
            amount = round(random.uniform(10, max_withdrawal_amount), 2)
        else:
            amount = round(random.uniform(10, 1000), 2)

        # Create the transaction record
        transaction = {
            "TransactionID": len(transactions) + 1,
            "AccountID": account['AccountID'],
            "TransactionType": transaction_type,
            "Amount": amount,
            "TransactionDate": transaction_date.strftime('%Y-%m-%d %H:%M:%S'),
        }

        # If it's a transfer transaction, create related transactions
        if transaction_type == 'Transfer':
            # Determine the source and target accounts (randomly)
            source_account = account['AccountID']
            target_account = random.choice(accounts_df['AccountID'])
            while target_account == source_account:
                # Ensure the source and target accounts are different
                target_account = random.choice(accounts_df['AccountID'])

            # Create a withdrawal transaction from the source account
            withdrawal_transaction = transaction.copy()
            withdrawal_transaction["Amount"] *= -1  # Make the amount negative
            withdrawal_transaction["TransactionType"] = 'Withdrawal'
            withdrawal_transaction["AccountID"] = source_account

            # Create a deposit transaction into the target account
            deposit_transaction = transaction.copy()
            deposit_transaction["AccountID"] = target_account

            # Add the withdrawal and deposit transactions to the list
            transactions.append(withdrawal_transaction)
            transactions.append(deposit_transaction)

            # Update the withdrawal transaction's RelatedTransactionID with the deposit transaction's ID
            withdrawal_transaction["RelatedTransactionID"] = deposit_transaction["TransactionID"]

            # Update the map to link the withdrawal transaction to the deposit transaction
            related_transactions[withdrawal_transaction["TransactionID"]] = deposit_transaction["TransactionID"]

        # Append the transaction to the list
        transactions.append(transaction)

# The 'transactions' list now contains multiple transaction records, including transfers and related transactions
transactions_df = pd.DataFrame(transactions)

```
